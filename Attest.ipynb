```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('kc-house-data.csv', encoding='latin-1')
df.head(15)
```

```python
plt.hist(df['price'])
plt.xlabel('Цена недвижимости')
plt.ylabel('Количество')
plt.title('Количество по цене')
```

**Из графика видно, что имеет место наибольшее количество квартир стоимостью до 1 миллиона рублей**

```python

plt.hist(df['sqft_living'])
plt.xlabel('Жилая площадь')
plt.ylabel('Количество')
plt.title('Количество по жилой площади')
```

**Из графика видно, что большее число недивижимости имеет квадратуру чуть больше 20 м^2**

```python
plt.hist(df['yr_built'])
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.title('Распределение по году постройки')
```

**Из графика видно, что наибольшее число недвижимости построено в течение первых двух десятилетий 21 века**

```python
data = df['waterfront'].value_counts()
data
```

```python
plt.pie(data.values, labels = ['Нет', 'Да'], autopct='%1.1f%%')
```

**Из графика видно, что подавляющее число недвижимости не имеет вида на набережную**

```python
data_for_floors = df['floors'].value_counts().sort_index()
data_for_floors
```

```python
df['floors'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Количество этажей')
plt.ylabel('Количество')
plt.title('Распределение этажей')
```

Из графика видно, что наибольшее количество недвижимости имеют один этаж

```python
data_con = df['condition'].value_counts()
data_con
```

```python
df['condition'].value_counts().sort_index().plot(kind='bar', rot = 0)
plt.xlabel('Оценка')
plt.ylabel('Количество')
plt.title('Состояние недвижимости')
```

Из графика видно, что большее количество недвижимости оценивается на удовлетворительно

```python
import seaborn as sns
sns.boxplot(x=df['bedrooms'], y=df['price'], showfliers=False)
```

Исходя из оценки графика, можно сделать вывод о наличии прямой зависимости иежду ценой и количеством комнат в недвижимости. Чем больше комнат, тем выше цена.

```python

sns.barplot(data=df, x='waterfront', y='price')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость недвижимости')
plt.title('Распределение стоимости недвижимости по виду на набережную')
```

На графике наглядно представлено, что наличие вида на набережную значительно повышает стоимость недвижимости.

```python
data = pd.crosstab(index=df['price'], columns=df['floors'])
data.reset_index()
data
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sqft_living'], df['price'], color='skyblue', alpha=0.5)
plt.title('Зависимость цены от площади недвижимости')
plt.xlabel('Площадь недвижимости')
plt.ylabel('Цена, $')
plt.show()
```

Также как и на предыдущем графике видно, что чем больше плозадь недвижимости, тем больше цена на неё.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['bathrooms'], df['price'], color='purple', alpha=0.5)
plt.title('Зависимость цены от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Цена, $')
plt.show()
```

Аналогично с предыдущими примерами имеем дело с прямой зависимостью, чем больше ванных комнат, тем выше стоимость.

```python
sns.boxplot(x=df['condition'], y=df['price'], showfliers=False)
```

Лучше условия, выше цена.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['grade'], df['price'], color='brown', alpha=0.5)
plt.title('Зависимость цены от оценки качества недвижимости')
plt.xlabel('Оценка качества недвижимости')
plt.ylabel('Цена, $')
plt.show()
```

Чем выше оценивается качество недвижимости, тем выше её цена.
